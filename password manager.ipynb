```python
import os

FILE_NAME = "password.txt"

def load_passwords():
    passwords= {}
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r") as file:         #(with open is used so the python automatically closes the file after reading)
                                                   #(opens the file in read mode("r"))

            for line in file:
                website, username, password = line.strip().split("|")
                passwords[website] = {"username": username, "password": password}
    return passwords

def save_passwords(passwords):                      #(defines a function to write passwords back to the file whenever we add new ones)
                                                    
    with open(FILE_NAME, "w") as file:              #(opens the file in write mode("w"))
        
        for website, data in passwords.items():
            file.write(f"{website}|{data['username']}|{data['password']}\n")    #(loops through each website in dictionary)
                                                                                #(\n ensures each entry is on a new line)

def add_password(passwords):
    website=input("Enter the website name: ").lower()
    username=input("Enter username: ")
    password=input("Enter password")

    passwords[website]={
        "username": username,                         #(add the new entry to the dictionary)
        "password": password
    }

    save_passwords(passwords)                         #(calls save_passwords() to update the file)
    print("password has been saved successfully!")
    
def get_password(passwords):
    website = input("Enter website name to search: ").lower()

    if website in passwords:
        print("\n Credentials Found: ")
        print("Username: ", passwords[website]["username"])
        print("Password: ", passwords[website]["password"], "\n")

    else:
        print("No password found for this website.\n")

def main():
    passwords = load_passwords()

    while True:
        print("====Password Manager====")
        print("1. Add a new password")
        print("2. Get a saved password")
        print("3. Exit")

        choice = input("Choose an option (1/2/3): ")

        if choice == "1":
            add_password(passwords)

        elif choice == "2":
            get_password(passwords)

        elif choice == "3":
            print("Exiting Password Manager. Bye.")
            break

        else:
            print("Invalid choice. Try again.\n")

main()
```
